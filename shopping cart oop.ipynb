```python
from IPython.display import clear_output

def shopping_cart():
    shopping_list = []
        
    done = False
    
    while not done:
        print("""
Welcome to your shopping cart!        
Type 'quit' to exit program and print receipt
Type 'view' to view items in current shopping cart
Type 'add' to add item to shopping cart
Type 'remove' to remove item from shopping cart
""")
        request = input()
        clear_output()
            
        if request == 'quit':
            done = True
            for i in shopping_list:
                print(i)
            
        elif request == 'view':
            for i in shopping_list:
                print(i)
                    
        elif request == 'add':
            addition = input("Add item to shopping cart: ")
            info = [addition]
            shopping_list.append(info)
                
        elif request == 'remove':
            subtraction = input("Remove item from shopping cart: ")
            info = [subtraction]
            shopping_list.remove(info)
                
        else:
            print("""
You have not typed in a valid request, please see the instructions below
""")
                
shopping_cart()
```

```python
from IPython.display import clear_output

class Cart:
    def __init__(self):
        self.items = []
        
    def add_item(self, obj):
        return self.items.append(obj)
    
    def remove_item(self, product_name):
        for p in self.items:
            if product_name in [obj.name for obj in self.items]:
                self.items.remove(p)
                
    def shopping_list(self):
        total = 0
        if not self.items:
            print('Your cart is currently empty')
        else:
            show_cart = []
            for p in self.items:
                if p.name not in [i['name'] for i in show_cart]:
                    show_cart.append({
                        'name': p.name,
                        'price': p.price
                    })
                
                else:
                    for i in show_cart:
                        if i['name'] == p.name:
                            i['price'] += p.price
            
            
            for i in range(len(show_cart)):
                print(f"{[i['name'] for i in show_cart].count(show_cart[i]['name'])} {show_cart[i]['name']} [${show_cart[i]['price']}]")
                
        print('<>' * 40)
                      
        total += float(sum([p.price for p in self.items]))
                      
        print(f"Total: ${total:,.2f}")
                      

class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price    
    
class shopping_cart:
    def show_instructions(self):
        print("""
Welcome to your shopping cart!        
Type 'quit' to exit program and print receipt
Type 'view' to view items in current shopping cart
Type 'add' to add item to shopping cart
Type 'remove' to remove item from shopping cart
""")
        
    @classmethod
    def run(self):
        cart = Cart()
                      
        done = False

        while not done:
            self.show_instructions(self)
            request = input()
            clear_output()

            if request == 'quit':
                done = True
                cart.shopping_list()

            elif request == 'view':
                cart.shopping_list()

            elif request == 'add':
                item_name = input("Add item to shopping cart: ")
                item_price = float(input("Enter the price of the object: "))
                cart_item = Item(item_name, item_price)
                cart.add_item(cart_item)

            elif request == 'remove':
                cart_item = input("Remove item from shopping cart: ")
                cart.remove_item(cart_item)

            else:
                print("""
    You have not typed in a valid request, please see the instructions below
    """)

shopping_cart.run()
```
